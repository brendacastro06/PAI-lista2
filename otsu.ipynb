```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# função para exibir imagens
def show_img(img, title="Imagem"):
    plt.figure(figsize=(6,6))
    if len(img.shape) == 2:  # grayscale
        plt.imshow(img, cmap="gray")
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
# função para exibir histograma
def hist_img(img, title="Imagem"):
    plt.subplot(1, 2, 2)
    plt.hist(img.ravel(), bins=256, range=[0, 256], color='gray')
    plt.title(title)
    plt.xlabel("Intensidade do Pixel")
    plt.ylabel("Número de Pixels")
    plt.xlim([-10, 265])
    plt.tight_layout()
    plt.show()
```

```python
# carregar a imagem
img = cv2.imread('pantera1.jpeg')
show_img(img, "Imagem Original")
```

```python
# converta para escala de cinza
img_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_img(img_cinza, "Imagem Original - Escala de Cinza")
hist_img(img_cinza, "Histograma Imagem em Escala de Cinza")
```

```python
# aplicando a Limiarização de Otsu
ret_otsu, img_otsu = cv2.threshold(img_cinza, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# exibe os resultados
show_img(img_otsu, f'Limiarização de Otsu (Limiar = {ret_otsu})')
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img_otsu, cmap="gray")
plt.axis("off")
plt.savefig(
    'pantera-otsu.jpeg',
    bbox_inches='tight',
    pad_inches=0
)
plt.show()
```
